```python
####
####### REINFORCEMENT LEARNING PROJECT 2024
####### INES & ISABELLA
####

# TOPIC:
# OPTIMIZATION OF THE TRAFFIC LIGHT SYSTEM
# Reference: Zeinaly, Z., Sojoodi, M., & Bolouki, S. (2023).
# A Resilient Intelligent Traffic Signal Control Scheme for Accident Scenario
# at Intersections via Deep Reinforcement Learning. *Sensors*, 23(3), 1578.

# This code is realised to optimize the traffic light system in a scenario
# simplified with respect the one of the paper since we have considered a
# four-way intersection with streets of one lane each that only allow
# vehicles to go straight on. It means that vehicles are only allowed
# to travel north-south, east-west in both directions.

# I STEP - Import the packages
import numpy as np
import random                        # to randomise vehicle arrivals and directions
import pandas as pd
from prettytable import PrettyTable  # for more pretty table
import matplotlib.pyplot as plt      # for plotting

# II STEP - set the parameters for the Q-learning algorithm
# We suppose that the MAX NUMBER OF VEHICLE PER LANE is 12 to simplify
# the context and obtain a computationally efficient problem to solve.
# There are TWO TRAFFIC PHASES: the green one for north-south direction and
# the green for east-west one.
# We set the LEARNING RATE alpha to 0.1 to make the q-value updates have
# a not too high impact on q-learning algorithm. In this way, the optimization
# process won't be so fast. This will grant stability in during the learning
# process by avoiding too big fluctuations.
# The DISCOUNT FACTOR (gamma) is set to 0.9 giving a quite important weight to
# future rewards and orienting the system to long-term strategy. An high
# discount factor enables the light system to consider the long-term consequences
# of his actions, encouraging decisions that reduce congestion on a permanent
# basis rather than temporarily.
# The EXPLORATION RATE (epsilon) is set to 1 to encourage exploration, allowing the agent
# to increase its experience and collect Q-values for more combinations of states
# and actions. It is very important to encourage exploration in a context where
# optimal behaviour must be identified.
# The EXPLORATION DECAY is a parameter that gradually reduces epsilon
# to each episode, encouraging the system to exploit learned strategies
# as the agent gains experience and finds effective actions.
# The NUMBER OF EPISODES is 1000, meaning that the model will repeat the
# learning cycle 1000 times. We set an high value to help the agent learning a
# robust strategy that will work in variable scenarios.
# The MAX NUMBER OF STEPS PER EPISODE defines the maximum duration of each
# learning episode. Setting 100 steps per episode enables to avoid exceeding
# the number of iterations, which would be computationally inefficient.
num_vehicles_per_lane = 12  # Maximum number of vehicles per lane
num_traffic_phases = 2      # Two traffic light phases: 0 = green for NS, 1 = green for EW
learning_rate = 0.1         # Learning rate (alpha)
discount_factor = 0.9       # Discount factor (gamma)
exploration_rate = 1.0      # Initial exploration rate (epsilon)
exploration_decay = 0.99    # Exploration decay rate
num_episodes = 10000        # Total number of episodes
max_steps_per_episode = 100 # Max steps per episode

# STEP 3 - DEFINE THE STATE, ACTIONS
# - STATE
# the State is defined as a vector of three numbers such as (3,5,0) explaining
# how many vehicles are waiting in the north-south direction (3), how many vehicles
# are waiting in the east-west direction (5) and which is the the green phase
# (NS, since it is 0). By generalizing, the State is defined by the vector
# (n*NS, n*EW, phase).
# - ACTION
# The actions the agent can execute are: "keep the green light",
# "switch the light". As decided before, the first action is
# identified by the 0, while the second by the 1.

# STEP 4 - INITIALIZE THE PROBLEM
# - DEFINE THE TOTAL STATES, NUMBER OF ACTIONS
# The total number of states is obtained by multiplying the allowed number of
# vehicles per lane in each direction by the number of traffic phases. In this
# way, we will consider all the possible combinations of number of cars in the
# four-way intersection. We notice that the problem can easily become
# computationally complex, but fortunately its complexity is not exponential.
# The total number of actions is two and do coincide with the number of traffic phases.
total_states = num_vehicles_per_lane * num_vehicles_per_lane * num_traffic_phases
num_actions = num_traffic_phases  # Actions: 0 = "keep phase", 1 = "switch phase"

# - DEFINE THE Q-TABLE AND INITIALIZE IT TO 0
# Initialize the Q table with zeros
# The Q-table dimension is given by the total number of states and the
# total number of actions since the Q(s,a) value - state, action value - is
# the Q-value is computed considering the combination of the actual action
# and actual state.
Q = np.zeros((total_states, num_actions))

# - DEFINE THE CUMULATIVE REWARD (Gt), NUMBER OF VEHICLE PER EPISODE
# Track cumulative rewards and the total number of vehicles per episode for plotting
cumulative_rewards = []
total_vehicles_per_episode = []

# - DEFINE THE Q-VALUES LIST
# To store Q-values over episodes for a specific state-action pair
q_values_over_time = []

# - DEFINE THE QUEUE LENGHTS PER EPISODE LIST
# Lists to track the queue lengths per episode
queue_ew_per_episode = []
queue_ns_per_episode = []

# STEP 4 - CREATE THE PREPARATORY FUNCTIONS TO SET THE Q-LEARNING ALGORITHM
# - DEFINE THE FUNCTION GET_STATE_INDEX
# This function is fundamental to describe with a scalar value the actual state
# in order to simplify the updating process of Q-value, since it enable to
# identify the current state no more with a vector, but with a scalar.
# It takes as inputs the parameters of the vector mapping the current state
# and applying some operations to get a scalar. In particular, it multiplies
# each combination of number of vehicles in the intersection
 # (num_vehicles_per_lane * num_traffic_phases) by the number of vechicles in
 # east-west direction. The resulting value is summed by the multiplication
 # between the number of vehicles in north south direction and the number of
 # traffic phases. Finally, the resulting value is summed by the current traffic
 # phase. These kinds of operations are defined in order to have a scalar
 # value that UNIQUELY IDENTIFIES each states. That's why it needs to take into
 # account the number of combination of number of vehicles waiting at the intersection.
def get_state_index(vehicles_ew, vehicles_ns, traffic_phase):
    return vehicles_ew * (num_vehicles_per_lane * num_traffic_phases) + vehicles_ns * num_traffic_phases + traffic_phase


# - DEFINE THE REWARD FUNCTION
# The reward is computed summing three components:
# 1. the waiting penalty, the reward obtained by the agent depending on the number of vehicles waiting at the intersection;
# 2. the improvement reward, that is given depending on the changed number of waiting cars;
# 3. the change penalty, that is set to avoid the agent switching too much.
# All the three components are computed on the basis of the actual and previous numbers of vehicles
# waiting at the intersection and the action executed by the agent.

def calculate_reward(vehicles_ew, vehicles_ns, prev_vehicles_ew, prev_vehicles_ns, traffic_phase, action):

    # 1° COMPONENT - waiting penalty based on active traffic phase
    # If the light is green for the direction North-South, it sets
    # the first component equal to 10 - the number of waiting vectors
    # in the North_south direction and the same for the other direction.
    if traffic_phase == 0:
        waiting_reward = (10 - vehicles_ns)  # North-South phase active
    else:
        waiting_reward = ( 10 - vehicles_ew)  # East-West phase active

    # 2° COMPONENT - improvement reward
    # It is computed the difference in the number of waiting cars in
    # both directions between the actual and previous state and it is
    # assigned a reward (improvement reward) given by the sum of those
    # resulting values. If there is a decrease in the number of waiting cars,
    # the improvement reward will be positive, it will be negative otherwise.
    # Improvement reward: Calculate the decrease in the number of waiting cars
    improvement_ns = (prev_vehicles_ns - vehicles_ns)  # Positive if NS vehicles decreased
    improvement_ew =  (prev_vehicles_ew - vehicles_ew)  # Positive if EW vehicles decreased
    improvement_reward = improvement_ns + improvement_ew      # Sum of improvements in both directions

    # 3° COMPONENT - change penalty
    # A small penalty of -1 is applied when the agent chooses to switch the light
    # in order to avoid it is changed too much.
    change_penalty = -1 if action == 1 else 0

    # TOTAL REWARD COMPUTATION
    # Total reward is a combination of the two penalties and the improvement reward
    reward = waiting_reward + change_penalty + improvement_reward
    return reward

# - DEFINE THE PERFORM_ACTION FUNCTION
# It is the function that encapsulates how the traffic system evolves based on
# the current state and the action taken by the agent. This function enables
# to move from the current state to another on the basis of the action chosen
# by the agent. To identify the characteristics of the next state, it is
# necessary to decompose the index corrisponding to the current state and
# get its component. So, at the beginning we just computed the opposite
# formula that we use to convert the vector state in a scalar vector to obtain
# again the number of the vehicles in the direction North-south, the number
# of vehicles in the other direction and the traffic phase.
# Then, based on which traffic phase is active, so if it is green the light
# for north-south or the opposite, the number of waiting vehicles in a
# direction reduces by 1 or 3, depending on how many vehicles manages to
# pass when the green phase is active.
# However, while some vehicles are passing the intersection, other ones arrives.
# The number of cars arriving is randomly determined. To get the new number
# of vehicles waiting at the light, we have to sum the number of waiting vehicles
# after some have pass the intersection during the green phase with the random
#  number of arriving cars. Since we have fixed the maximum number of vehicles
# per lane equal to 12, we must be sure that, in the next state, this value is
# not exceeded. That's why the vehicles number is computing as a minimum between
# allowed number of vehicles per lane (num_vehicles_per_lane - 1) and the
# sum between the waiting cars and the new ones arriving
# (vehicles_ew + random.randint(0, 2).
# The allowed number of vehicles per lane is computed as
# num_vehicles_per_lane - 1 because in the system starts to count from 0.
# So, it is just a mathematical reason.
# Then, if the agent has chosen to switch the phase, we have to switch the traffic
# phase from 1 to 0 or viceversa. The function for doing it is traffic_phase = 1 - traffic_phase.
# Finally, we have to convert the new state into a scalar that uniquely identifies it.
# To do that we have use the same formula described in the get_state_index function.
def perform_action(current_state, action):
    vehicles_ew = current_state // (num_vehicles_per_lane * num_traffic_phases)
    vehicles_ns = (current_state // num_traffic_phases) % num_vehicles_per_lane
    traffic_phase = current_state % num_traffic_phases

    # Adjust vehicle counts based on the traffic phase
    if traffic_phase == 0:  # Green for NS
        vehicles_ns = max(0, vehicles_ns - random.randint(0, 3))
    else:  # Green for EW
        vehicles_ew = max(0, vehicles_ew - random.randint(0, 3))

    # Randomly increase vehicle counts to simulate new arrivals
    vehicles_ew = min(num_vehicles_per_lane - 1, vehicles_ew + random.randint(0, 2))
    vehicles_ns = min(num_vehicles_per_lane - 1, vehicles_ns + random.randint(0, 2))

    # Switch phase if action is 1
    if action == 1:
        traffic_phase = 1 - traffic_phase

    return get_state_index(vehicles_ew, vehicles_ns, traffic_phase)

# - DEFINE THE Q-TABLE
# The q-table is initialised by defining the columns names "States", "Keep Phase",
#  "Switch Phase", indicating that the first column will contain the states and
# and the other two the q-values depending on the state and the action chosen by
# the agent. To account for all the possible states in the first column, the
# the function iterates over all possible vehicle counts for both the east-west
# (ew) and north-south (ns) directions, as well as the two traffic phases.
# For each combination of these values, it calculates a state index using
# the get_state_index() function. Then, for each state, it retrieves the
# Q-values associated with the two possible actions (keeping the phase or
# switching the phase) from the Q-table, formats them to four decimal places,
# and adds them to the table. Finally, the table is printed, showing the
# Q-values for all possible states and actions. In the table the states are not
# visualize through the index, but through its vector components, so, by declaring
# the number of waiting car in east-west direction, the number of waiting cars
# in the other one and the actual phase.

def print_pretty_q_table(Q):
    # Initialize PrettyTable
    table = PrettyTable()
    table.field_names = ["States", "Keep Phase", "Switch Phase"]

    # Generate index based on state structure and fill the table
    for ew in range(num_vehicles_per_lane):
        for ns in range(num_vehicles_per_lane):
            for phase in range(num_traffic_phases):
                state_index = get_state_index(ew, ns, phase)
                # Format Q-values to four decimal places
                keep_phase_value = f"{Q[state_index][0]:.4f}"
                switch_phase_value = f"{Q[state_index][1]:.4f}"
                table.add_row([f"EW:{ew} NS:{ns} Phase:{phase}", keep_phase_value, switch_phase_value])

    # Print the table with a clear title
    print("\nQ-Table:")
    print("Table with 'States' as the index and actions ('Keep Phase' and 'Switch Phase') as columns:\n")
    print(table)

# - DEFINE THE Q-LEARNING TRAINING LOOP
# For each of the 1000 episodes, the initial values are set. Firstly, the initial
# parameters of the State are randomly determined as a value between 0 and the
# allowed number of vehicles per lane. Then, the initial state is converted
# into the index identifying the current state. The other values that must be
# initialized are total_vehicles, that counts the total number of waiting vehicles
# in both directions for episode, the ew_queue and ns_queue, that store the
# number of waiting vehicles respectively in direction east-west and north-south.
# Once the variables are initialized, the action loop can be defined. It states
# that the loop continues as long as the episode is not marked as done and
# the number of steps does not exceed the limit (max_steps_per_episode).
# Then, an action must be chosen. So, firstly, the state components are extracted
# as usual and then an action is selected on the bases of them. In particulare,
# the action is selected with the epsilon-greedy algorith, according to which,
# the agent chooses a random action with epsilon probability and a greedy action
# with 1-epsilon probability. This algorithm encourages the agent to explore new
# actions initially and gradually to explot the learned Q-values over time thanks
# to the decayed exploration rate.
# Once the action is chosen, it is performed and a new state is obtained by
# applying the previously defined perform_action to the current state and the
# current action. Since we want to keep track of the new number of vehicles in
# each direction in the new state, we extract those values from the scalar that
# describes this new state.
# The action generate a reward that it computed for the current step based on the
# transition from the previous to the new state.
# Finally, the Q-table is computed using the Q-learning update rule, called
# Bellman equation. The equation updates the Q-value based on the reward and
# the maximum possible future reward (best_next_action).
# After that, we update all the tracking variables that will be useful for
# plotting and store various performance metrics such as the cumulative_rewards,
# total_vehicles_per_episode, q_values_over_time and so on.
# At the end of the episode, the exploration rate is gradually decreased by
# the exploration_decay, encouraging the agent to exploit its learned
# knowledge more as the episodes progress.
for episode in range(num_episodes):
    initial_ew = random.randint(0, num_vehicles_per_lane - 1)
    initial_ns = random.randint(0, num_vehicles_per_lane - 1)
    initial_phase = random.randint(0, 1)

    current_state = get_state_index(initial_ew, initial_ns, initial_phase)
    done = False        # to check if the episode is finished
    step_count = 0      # to limit steps per episode
    episode_reward = 0  # track reward for this episode

    # Initialize total vehicles for the episode
    total_vehicles = initial_ew + initial_ns

    # Initialize queue tracking for the episode
    ew_queue = initial_ew
    ns_queue = initial_ns

    while not done and step_count < max_steps_per_episode:
        prev_vehicles_ew = current_state // (num_vehicles_per_lane * num_traffic_phases)
        prev_vehicles_ns = (current_state // num_traffic_phases) % num_vehicles_per_lane
        prev_traffic_phase = current_state % num_traffic_phases

        if random.uniform(0, 1) < exploration_rate:
            action = random.choice([0, 1])  # Random action
        else:
            action = np.argmax(Q[current_state])  # Best action based on Q-values

        new_state = perform_action(current_state, action)
        # If needed, you can calculate new_ew and new_ns from new_state
        new_ew = new_state // (num_vehicles_per_lane * num_traffic_phases)
        new_ns = (new_state // num_traffic_phases) % num_vehicles_per_lane
        reward = calculate_reward(
            new_state // (num_vehicles_per_lane * num_traffic_phases),
            (new_state // num_traffic_phases) % num_vehicles_per_lane,
            prev_vehicles_ew,
            prev_vehicles_ns,
            prev_traffic_phase,
            action
        )

        # Update episode reward
        episode_reward += reward

        # Update total vehicles for this step (this is to track it within the episode)
        total_vehicles = (new_state // (num_vehicles_per_lane * num_traffic_phases)) + ((new_state // num_traffic_phases) % num_vehicles_per_lane)

       # Update Q-table using the Q-learning formula
        best_next_action = np.max(Q[new_state])
        Q[current_state, action] += learning_rate * (reward + discount_factor * best_next_action - Q[current_state, action])

        # Update queues for this episode
        ew_queue = new_ew
        ns_queue = new_ns

        current_state = new_state
        step_count += 1

    # Store the cumulative reward for this episode
    cumulative_rewards.append(episode_reward)

    # Store the total vehicles for this episode after the loop
    total_vehicles_per_episode.append(total_vehicles)  # Append total vehicles to the list

    # Store the queue lengths for this episode
    queue_ew_per_episode.append(ew_queue)
    queue_ns_per_episode.append(ns_queue)

    # Store the Q-values for the current state after completing the episode
    q_values_over_time.append(Q[current_state].copy())  # Record Q-values for the last state of the episode

    # Decay the exploration rate
    exploration_rate *= exploration_decay

print("Q-Learning complete!")
print_pretty_q_table(Q)

# Convert to NumPy array for easier handling and plotting
q_values_over_time = np.array(q_values_over_time)  # Convert to a NumPy array for easier handling


# STEP 5 - VISUALIZATION OF THE RESULTS
# To better understand the performance of our model in solving the problem of
# flow of cars optimization, we can create some plots.
# To better identify the trend, we add to the graphs a curve
# indicating the smoother moving average.
# - Define the MOVING AVERAGE
# We set a larger moving average window size for smoother lines and then we
# write the function that creates the moving average. More smoothness is
# obtained by increasing the window_size.
window_size = 100
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size) / window_size, mode='valid')


# PLOT 1 - CUMULATIVE REWARDS OVER EPISODES WITH A SMOOTHER MOVING AVERAGE
# Realise the plot
# At first the plot is initialised by setting its dimension of 10 units wide by
# 5 units tall, providing enough space to clearly display data for multiple episodes.
# The plot will show the cumulative rewards on the y axis and the episode number
# on the x axis. Then, the smoothed moving average is added to the plot.
plt.figure(figsize=(10, 5))
plt.plot(range(num_episodes), cumulative_rewards, label='Cumulative Reward per Episode')
plt.plot(range(window_size - 1, num_episodes), moving_average(cumulative_rewards, window_size), color='red', linestyle='-', linewidth=2, label='Smoothed Moving Average')
plt.xlim(0, 600)  # Limit x-axis to the first 600 episodes
plt.xlabel('Episodes')
plt.ylabel('Cumulative Reward (G_t)')
plt.title('Figure 1 : Cumulative Reward Over Episodes in Q-Learning')
plt.legend()
plt.grid()
plt.show()
# COMMENT: This graph shows that the cumulative reward increases with the episodes,
# indicating a positive relationship between the cumulative reward
# and the number of episode. The result is good indicating that the agent,
# with its choices is effectively perceiving the objective of gaining the
# highest possibile cumulative discounted reward.


# PLOT 2 - TOTAL NUMBER OF VEHICLES OVER EPISODES WITH A SMOOTHER MOVING AVERAGE
# Realise the plot
# At first the plot is initialised by setting its dimension of 10 units wide by
# 5 units tall, providing enough space to clearly display data for multiple episodes.
# The plot will show the total number of vehicles per episode on the y axis
# and the episode number on the x axis. Then, the smoothed moving average
# is added to the plot.
plt.figure(figsize=(10, 5))
plt.plot(range(num_episodes), total_vehicles_per_episode, label='Total Vehicles (NS + EW)', color='orange')
plt.plot(range(window_size - 1, num_episodes), moving_average(total_vehicles_per_episode, window_size), color='red', linestyle='-', linewidth=2, label='Smoothed Moving Average')
plt.xlim(0, 600)  # Limit x-axis to the first 600 episodes
plt.xlabel('Episodes')
plt.ylabel('Total Number of Vehicles')
plt.title('Figure 2 : Total Vehicles Over Episodes in Q-Learning')
plt.legend()
plt.grid()
plt.show()
# COMMENT: This graph shows that the cumulative reward decreases with the
# episodes, indicating a positive relationship between the total number of
# vehicles arriving and the number of episodes. The result is good indicating
# that the agent, with its choices is effectively reducing the total number
# of cars waiting at the intersection. However, if we look at the orange line,
# we notice that obviously randomness is still there, since the vehicles
# randomly arrive to the intersection and a random number of vehicles pass
# through it.


# PLOT 3 - Q-VALUES OVER TIME FOR A SPECIFIC STATE-ACTION PAIR WITH A SMOOTHER
# Realise the plot
# At first the plot is initialised by setting its dimension of 10 units wide by
# 5 units tall, providing enough space to clearly display data for multiple episodes.
# The plot will show the q-values over time on the y axis
# and the episode number on the x axis. Then, the smoothed moving average
# is added to the plot. To focus only on the first 600 episodes, we zooom
# the plot limiting the x axis in the [0,600] interval.
plt.figure(figsize=(10, 5))
plt.plot(range(len(q_values_over_time)), q_values_over_time[:, 0], label='Q-value for Keep Phase (Action 0)', color='blue')
plt.plot(range(len(q_values_over_time)), q_values_over_time[:, 1], label='Q-value for Switch Phase (Action 1)', color='orange')
plt.plot(range(window_size - 1, len(q_values_over_time)), moving_average(q_values_over_time[:, 0], window_size), color='red', linestyle='-', linewidth=2, label='Smoothed Moving Average for Keep Phase')
plt.plot(range(window_size - 1, len(q_values_over_time)), moving_average(q_values_over_time[:, 1], window_size), color='red', linestyle='--', linewidth=2, label='Smoothed Moving Average for Switch Phase')
plt.xlim(0, 600)  # Limit x-axis to the first 600 episodes
plt.xlabel('Episodes')
plt.ylabel('Q-value')
plt.title('Figure 3 : Q-values Over Time for Specific State-Action Pair')
plt.legend()
plt.grid()
plt.show()
# COMMENT: This graph shows two q-values curves, one for each of the action
# the agent can choose. Both curves are increasing at a decreasing rate,
# meaning that in the first episodes, the agent explores and progressively its
# its choices. Then, the agent will exploit stabilizing its choices since he
# will continue to choose the optimal ones. This means that the agent is
# effectively learning the optimal strategy and uses it. In other words, the
# Q-values are effectively converging to optimal values.
# We notice that one curve reaches higher q-values than the other one. It happens
# because of the change_penalty we applied.


# PLOT 4 - SCATTER PLOT BETWEEN EW and NS QUEUES
# Realise the plot
plt.figure(figsize=(10, 6))
plt.scatter(queue_ew_per_episode, queue_ns_per_episode, alpha=0.6, c='green')
plt.xlabel('Queue Length in EW Direction')
plt.ylabel('Queue Length in NS Direction')
plt.title('Figure 4 : Substitution Phenomenon Between EW and NS Queues')
plt.grid(True)
plt.show()
# COMMENT:
# This plot shows the "Substitution Phenomenon" between East-West (EW) and
# North-South (NS) queues, where a longer queue in one direction generally
# corresponds to a shorter queue in the other, indicating that a substitution
# effect is verifying. Darker green points indicate more frequent queue
# length combinations, with clustering around specific values.
# This suggests that the traffic signal often causes one direction
# to clear while the other builds up, highlighting an inverse
# relationship that could inform better traffic light timing to
# balance queues and reduce congestion.

```

```python
# STEP 6 - VARIATION TO THE INITIAL PROBLEM
# Until now, we have approached the problem of the traffic flow optimization
# managed by a traffic light system considering that the vehicles randomly arrive
# at the intersection following the a uniform distribution. This means that the
# number of vehicles tends to fluctuates around a mean value, with main of the
# observations close to this central value. The probability that a very big or
# a very small number of vehicles arrive at the intersection is low. So, in this
# case, the traffic is moderate and foreseeable for what we discussed before.
# There will be occasional peaks and drops in the number of vehicles, but
# they will be rare.
# However, this is not what happens in the reality. The traffic is not
# foreseeable, there can be patterns in the days, but the traffic is not always
# moderate. There are some peaks, some drops and they occur frequently.
# Therefore, we will apply again the same model by switching the distribution
# of vehicles from the uniform to the poisson distribution that better describes
# the real flow of vehicles in the street.

# We run the same code of before but specifying ...
# instead of

# I STEP - Import the packages
import numpy as np
import random                        # to randomise vehicle arrivals and directions
import pandas as pd
from prettytable import PrettyTable  # for more pretty table
import matplotlib.pyplot as plt      # for plotting

# II STEP - set the parameters for the Q-learning algorithm
num_vehicles_per_lane = 12  # Maximum number of vehicles per lane
num_traffic_phases = 2      # Two traffic light phases: 0 = green for NS, 1 = green for EW
learning_rate = 0.1         # Learning rate (alpha)
discount_factor = 0.9       # Discount factor (gamma)
exploration_rate = 1.0      # Initial exploration rate (epsilon)
exploration_decay = 0.99    # Exploration decay rate
num_episodes = 30000        # Total number of episodes
max_steps_per_episode = 100 # Max steps per episode

# STEP 3 - DEFINE THE STATE, ACTIONS
total_states = num_vehicles_per_lane * num_vehicles_per_lane * num_traffic_phases
num_actions = num_traffic_phases  # Actions: 0 = "keep phase", 1 = "switch phase"

# - DEFINE THE Q-TABLE AND INITIALIZE IT TO 0
# Initialize the Q table with zeros
Q = np.zeros((total_states, num_actions))

# - DEFINE THE CUMULATIVE REWARD (Gt), NUMBER OF VEHICLE PER EPISODE
# Track cumulative rewards and the total number of vehicles per episode for plotting
cumulative_rewards = []
total_vehicles_per_episode = []

# - DEFINE THE Q-VALUES LIST
# To store Q-values over episodes for a specific state-action pair
q_values_over_time = []

# - DEFINE THE QUEUE LENGHTS PER EPISODE LIST
# Lists to track the queue lengths per episode
queue_ew_per_episode = []
queue_ns_per_episode = []

# STEP 4 - CREATE THE PREPARATORY FUNCTIONS TO SET THE Q-LEARNING ALGORITHM
# - DEFINE THE FUNCTION GET_STATE_INDEX
def get_state_index(vehicles_ew, vehicles_ns, traffic_phase):
    return vehicles_ew * (num_vehicles_per_lane * num_traffic_phases) + vehicles_ns * num_traffic_phases + traffic_phase


# - DEFINE THE REWARD FUNCTION
# The reward is computed summing three components:
# 1. the waiting penalty, the reward obtained by the agent depending on the number of vehicles waiting at the intersection;
# 2. the improvement reward, that is given depending on the changed number of waiting cars;
# 3. the change penalty, that is set to avoid the agent switching too much.
# All the three components are computed on the basis of the actual and previous numbers of vehicles
# waiting at the intersection and the action executed by the agent.

def calculate_reward(vehicles_ew, vehicles_ns, prev_vehicles_ew, prev_vehicles_ns, traffic_phase, action):

    # 1° COMPONENT - waiting penalty based on active traffic phase
    if traffic_phase == 0:
        waiting_reward = (10 - vehicles_ns)  # North-South phase active
    else:
        waiting_reward = ( 10 - vehicles_ew)  # East-West phase active

    # 2° COMPONENT - improvement reward
    improvement_ns = (prev_vehicles_ns - vehicles_ns)  # Positive if NS vehicles decreased
    improvement_ew =  (prev_vehicles_ew - vehicles_ew)  # Positive if EW vehicles decreased
    improvement_reward = improvement_ns + improvement_ew      # Sum of improvements in both directions

    # 3° COMPONENT - change penalty
    change_penalty = -1 if action == 1 else 0

    # TOTAL REWARD COMPUTATION
    # Total reward is a combination of the two penalties and the improvement reward
    reward = waiting_reward + change_penalty + improvement_reward
    return reward

# - DEFINE THE PERFORM_ACTION FUNCTION
# It is the function that encapsulates how the traffic system evolves based on
# the current state and the action taken by the agent. This function enables
# to move from the current state to another on the basis of the action chosen
# by the agent.
# Here there's the modification.
# - SET THE POISSON DISTRIBUTION
# Define Poisson arrival rate (mean number of vehicles arriving per time step)
# This setup should simulate a more realistic traffic flow with periods
#  of higher and lower vehicle arrivals, depending on the Poisson
# distribution's variance.
lambda_arrival_ew = 1.5  # mean arrival rate for EW direction
lambda_arrival_ns = 1.5  # mean arrival rate for NS direction

def perform_action(current_state, action):
    vehicles_ew = current_state // (num_vehicles_per_lane * num_traffic_phases)
    vehicles_ns = (current_state // num_traffic_phases) % num_vehicles_per_lane
    traffic_phase = current_state % num_traffic_phases

    # Adjust vehicle counts based on the traffic phase
    if traffic_phase == 0:  # Green for NS
        vehicles_ns = max(0, vehicles_ns - np.random.poisson(lambda_arrival_ew))
    else:  # Green for EW
        vehicles_ew = max(0, vehicles_ew - np.random.poisson(lambda_arrival_ns))

    # Randomly increase vehicle counts to simulate new arrivals
    vehicles_ew = min(num_vehicles_per_lane - 1, vehicles_ew + random.randint(0, 2))
    vehicles_ns = min(num_vehicles_per_lane - 1, vehicles_ns + random.randint(0, 2))

    # Switch phase if action is 1
    if action == 1:
        traffic_phase = 1 - traffic_phase

    return get_state_index(vehicles_ew, vehicles_ns, traffic_phase)

# - DEFINE THE Q-TABLE
# The q-table shows the Q-values for all possible states and actions.
# In the it, the states are not visualized through the index, but through
# its vector components, so, by declaring the number of waiting car in
# east-west direction, the number of waiting cars
# in the other one and the actual phase.

def print_pretty_q_table(Q):
    # Initialize PrettyTable
    table = PrettyTable()
    table.field_names = ["States", "Keep Phase", "Switch Phase"]

    # Generate index based on state structure and fill the table
    for ew in range(num_vehicles_per_lane):
        for ns in range(num_vehicles_per_lane):
            for phase in range(num_traffic_phases):
                state_index = get_state_index(ew, ns, phase)
                # Format Q-values to four decimal places
                keep_phase_value = f"{Q[state_index][0]:.4f}"
                switch_phase_value = f"{Q[state_index][1]:.4f}"
                table.add_row([f"EW:{ew} NS:{ns} Phase:{phase}", keep_phase_value, switch_phase_value])

    # Print the table with a clear title
    print("\nQ-Table:")
    print("Table with 'States' as the index and actions ('Keep Phase' and 'Switch Phase') as columns:\n")
    print(table)

# - DEFINE THE Q-LEARNING TRAINING LOOP
# For each of the 1000 episodes, the initial values are set. Firstly, the initial
# parameters of the State are randomly determined as a value between 0 and the
# allowed number of vehicles per lane. Then, the initial state is converted
# into the index identifying the current state. Some performance metrics and
# variables are stored and then the action is selected on the bases of state
# features, applying the epsilon-greedy algorith. Once the action is chosen,
# it is performed, a new state is reached and a reward is generated.
# Finally, the Q-table is computed using the Q-learning update rule, called
# Bellman equation and also the performance metrics and the interesting variables
# are stored again. At the end of the episode, the exploration rate is gradually decreased by
# the exploration_decay, encouraging the agent to exploit its learned
# knowledge more as the episodes progress.
for episode in range(num_episodes):
    initial_ew = random.randint(0, num_vehicles_per_lane - 1)
    initial_ns = random.randint(0, num_vehicles_per_lane - 1)
    initial_phase = random.randint(0, 1)

    current_state = get_state_index(initial_ew, initial_ns, initial_phase)
    done = False        # to check if the episode is finished
    step_count = 0      # to limit steps per episode
    episode_reward = 0  # track reward for this episode

    # Initialize total vehicles for the episode
    total_vehicles = initial_ew + initial_ns

    # Initialize queue tracking for the episode
    ew_queue = initial_ew
    ns_queue = initial_ns

    while not done and step_count < max_steps_per_episode:
        prev_vehicles_ew = current_state // (num_vehicles_per_lane * num_traffic_phases)
        prev_vehicles_ns = (current_state // num_traffic_phases) % num_vehicles_per_lane
        prev_traffic_phase = current_state % num_traffic_phases

        if random.uniform(0, 1) < exploration_rate:
            action = random.choice([0, 1])  # Random action
        else:
            action = np.argmax(Q[current_state])  # Best action based on Q-values

        new_state = perform_action(current_state, action)
        # If needed, you can calculate new_ew and new_ns from new_state
        new_ew = new_state // (num_vehicles_per_lane * num_traffic_phases)
        new_ns = (new_state // num_traffic_phases) % num_vehicles_per_lane
        reward = calculate_reward(
            new_state // (num_vehicles_per_lane * num_traffic_phases),
            (new_state // num_traffic_phases) % num_vehicles_per_lane,
            prev_vehicles_ew,
            prev_vehicles_ns,
            prev_traffic_phase,
            action
        )

        # Update episode reward
        episode_reward += reward

        # Update total vehicles for this step (this is to track it within the episode)
        total_vehicles = (new_state // (num_vehicles_per_lane * num_traffic_phases)) + ((new_state // num_traffic_phases) % num_vehicles_per_lane)

       # Update Q-table using the Q-learning formula
        best_next_action = np.max(Q[new_state])
        Q[current_state, action] += learning_rate * (reward + discount_factor * best_next_action - Q[current_state, action])

        # Update queues for this episode
        ew_queue = new_ew
        ns_queue = new_ns

        current_state = new_state
        step_count += 1

    # Store the cumulative reward for this episode
    cumulative_rewards.append(episode_reward)

    # Store the total vehicles for this episode after the loop
    total_vehicles_per_episode.append(total_vehicles)  # Append total vehicles to the list

    # Store the queue lengths for this episode
    queue_ew_per_episode.append(ew_queue)
    queue_ns_per_episode.append(ns_queue)

    # Store the Q-values for the current state after completing the episode
    q_values_over_time.append(Q[current_state].copy())  # Record Q-values for the last state of the episode

    # Decay the exploration rate
    exploration_rate *= exploration_decay

print("Q-Learning complete!")
print_pretty_q_table(Q)

# Convert to NumPy array for easier handling and plotting
q_values_over_time = np.array(q_values_over_time)  # Convert to a NumPy array for easier handling


# STEP 5 - VISUALIZATION OF THE RESULTS
# To better understand the performance of our model in solving the problem of
# flow of cars optimization, we can create some plots.
# To better identify the trend, we add to the graphs a curve
# indicating the smoother moving average.
# - Define the MOVING AVERAGE
#  More smoothness is obtained by increasing the window_size.
window_size = 100
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size) / window_size, mode='valid')


# PLOT 1 - CUMULATIVE REWARDS OVER EPISODES WITH A SMOOTHER MOVING AVERAGE
# Realise the plot
plt.figure(figsize=(10, 5))
plt.plot(range(num_episodes), cumulative_rewards, label='Cumulative Reward per Episode')
plt.plot(range(window_size - 1, num_episodes), moving_average(cumulative_rewards, window_size), color='red', linestyle='-', linewidth=2, label='Smoothed Moving Average')
plt.xlim(0, 600)  # Limit x-axis to the first 600 episodes
plt.xlabel('Episodes')
plt.ylabel('Cumulative Reward (G_t)')
plt.title('Figure 5 : Cumulative Reward Over Episodes in Q-Learning')
plt.legend()
plt.grid()
plt.show()
# COMMENT: This graph shows that the cumulative reward increases with the episodes,
# indicating a positive relationship between the cumulative reward
# and the number of episode. The result is good indicating that the agent,
# with its choices is effectively perceiving the objective of gaining the
# highest possibile cumulative discounted reward.


# PLOT 2 - TOTAL NUMBER OF VEHICLES OVER EPISODES WITH A SMOOTHER MOVING AVERAGE
# Realise the plot
plt.figure(figsize=(10, 5))
plt.plot(range(num_episodes), total_vehicles_per_episode, label='Total Vehicles (NS + EW)', color='orange')
plt.plot(range(window_size - 1, num_episodes), moving_average(total_vehicles_per_episode, window_size), color='red', linestyle='-', linewidth=2, label='Smoothed Moving Average')
plt.xlim(0, 600)  # Limit x-axis to the first 600 episodes
plt.xlabel('Episodes')
plt.ylabel('Total Number of Vehicles')
plt.title('Figure 6 : Total Vehicles Over Episodes in Q-Learning')
plt.legend()
plt.grid()
plt.show()
# COMMENT: This graph shows that the cumulative reward decreases with the
# episodes, indicating a positive relationship between the total number of
# vehicles arriving and the number of episodes. The result is good indicating
# that the agent, with its choices is effectively reducing the total number
# of cars waiting at the intersection. However, if we look at the orange line,
# we notice that obviously randomness is still there, since the vehicles
# randomly arrive to the intersection and a random number of vehicles pass
# through it.


# PLOT 3 - Q-VALUES OVER TIME FOR A SPECIFIC STATE-ACTION PAIR WITH A SMOOTHER
# Realise the plot
plt.figure(figsize=(10, 5))
plt.plot(range(len(q_values_over_time)), q_values_over_time[:, 0], label='Q-value for Keep Phase (Action 0)', color='blue')
plt.plot(range(len(q_values_over_time)), q_values_over_time[:, 1], label='Q-value for Switch Phase (Action 1)', color='orange')
plt.plot(range(window_size - 1, len(q_values_over_time)), moving_average(q_values_over_time[:, 0], window_size), color='red', linestyle='-', linewidth=2, label='Smoothed Moving Average for Keep Phase')
plt.plot(range(window_size - 1, len(q_values_over_time)), moving_average(q_values_over_time[:, 1], window_size), color='red', linestyle='--', linewidth=2, label='Smoothed Moving Average for Switch Phase')
plt.xlim(0, 600)  # Limit x-axis to the first 600 episodes
plt.xlabel('Episodes')
plt.ylabel('Q-value')
plt.title('Figure 7 : Q-values Over Time for Specific State-Action Pair')
plt.legend()
plt.grid()
plt.show()
# COMMENT: This graph shows two q-values curves, one for each of the action
# the agent can choose. Both curves are increasing at a decreasing rate,
# meaning that in the first episodes, the agent explores and progressively its
# its choices. Then, the agent will exploit stabilizing its choices since he
# will continue to choose the optimal ones. This means that the agent is
# effectively learning the optimal strategy and uses it. In other words, the
# Q-values are effectively converging to optimal values.


# PLOT 4 - SCATTER PLOT BETWEEN EW and NS QUEUES
# Realise the plot
plt.figure(figsize=(10, 6))
plt.scatter(queue_ew_per_episode, queue_ns_per_episode, alpha=0.6, c='green')
plt.xlabel('Queue Length in EW Direction')
plt.ylabel('Queue Length in NS Direction')
plt.title('Figure 8 : Substitution Phenomenon Between EW and NS Queues')
plt.grid(True)
plt.show()
# COMMENT:
# This plot shows the "Substitution Phenomenon" between East-West (EW) and
# North-South (NS) queues, where a longer queue in one direction generally
# corresponds to a shorter queue in the other, indicating that a substitution
# effect is verifying. Darker green points indicate more frequent queue
# length combinations, with clustering around specific values.
# This suggests that the traffic signal often causes one direction
# to clear while the other builds up, highlighting an inverse
# relationship that could inform better traffic light timing to
# balance queues and reduce congestion.




```

```python
#DEEP Q LEARNING ALGORITHM
#We are implementing a Deep Q-Learning (DQN) algorithm to optimize traffic light control at an intersection.
#The agent observes the number of vehicles in two directions (East-West and North-South) and the current traffic light phase.
#It then selects actions to either switch or maintain the traffic light phase.
#The goal is to minimize traffic congestion and waiting times by learning the best sequence of traffic light changes.
#The agent learns through interactions with the environment, receiving rewards based on the traffic flow efficiency.
#This approach uses deep neural networks to approximate the Q-values for state-action pairs.

import numpy as np               # For numerical operations, especially for array manipulation
import random                    # For generating random numbers for initial actions or states
import tensorflow as tf          # Framework for building and training neural networks
from tensorflow import keras     # High-level neural network API
from tensorflow.keras import layers  # Used to define different layers in a neural network model
from collections import deque     # Data structure to efficiently store and sample from memory

# Q-learning parameters
learning_rate = 0.1         # Alpha: controls the speed of learning updates
discount_factor = 0.9       # Gamma: balances importance of immediate vs. future rewards
exploration_rate = 1.0      # Epsilon: controls exploration vs. exploitation in action selection
exploration_decay = 0.99    # Decay rate for epsilon to reduce exploration over time
num_episodes = 10000        # Total learning episodes
max_steps_per_episode = 100 # Max actions per episode
batch_size = 32             # Number of experiences used in each training step

# Environment parameters
num_vehicles_per_lane = 12              # Max number of vehicles per lane (EW or NS)
num_traffic_phases = 2                  # Two possible traffic light phases: EW green or NS green
total_states = num_vehicles_per_lane * num_vehicles_per_lane * num_traffic_phases  # Total states in the environment
num_actions = num_traffic_phases       # Actions: 0 = keep current phase, 1 = switch phase

# Neural Network for Deep Q-Learning
class DQNAgent:
    def __init__(self, total_states, num_actions):
        self.total_states = total_states  # Total possible states
        self.num_actions = num_actions    # Total possible actions
        self.memory = deque(maxlen=2000)   # Replay memory to store past experiences
        self.gamma = discount_factor      # Discount factor for future rewards
        self.epsilon = exploration_rate   # Exploration rate
        self.epsilon_decay = exploration_decay  # Decay of exploration rate
        self.epsilon_min = 0.01           # Minimum exploration rate
        self.model = self.create_model()  # Main model for Q-learning
        self.target_model = self.create_model()  # Target model (for stability)
        self.update_target_model()        # Initialize the target model with the main model

    # Create the neural network model
    def create_model(self):
        model = keras.Sequential()
        model.add(layers.Dense(24, activation='relu', input_dim=self.total_states))  # First hidden layer
        model.add(layers.Dense(24, activation='relu'))  # Second hidden layer
        model.add(layers.Dense(self.num_actions, activation='linear'))  # Output layer (Q-values for each action)
        model.compile(optimizer=keras.optimizers.Adam(learning_rate=learning_rate), loss='mse')  # Compile model with mean squared error loss
        return model

    # Choose an action based on the current state (with exploration vs exploitation)
    def act(self, state):
        state = np.reshape(state, [1, self.total_states])  # Reshape state for model input
        if np.random.rand() <= self.epsilon:  # Exploration: random action
            return random.choice(range(self.num_actions))
        q_values = self.model.predict(state)  # Exploitation: choose action with highest Q-value
        return np.argmax(q_values[0])  # Return the index of the action with the max Q-value

    # Store experiences in memory
    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))  # Add experience to memory

    # Train the model using a batch of experiences sampled from memory
    def replay(self, batch_size):
        if len(self.memory) < batch_size:  # If there's not enough memory, return
            return
        minibatch = random.sample(self.memory, batch_size)  # Sample a batch of experiences
        for state, action, reward, next_state, done in minibatch:
            target = reward  # Initial target is the immediate reward
            if not done:  # If episode is not done, calculate future reward
                target += self.gamma * np.amax(self.target_model.predict(next_state)[0])  # Q-value for next state
            target_f = self.model.predict(state)  # Get current Q-values for the state
            target_f[0][action] = target  # Update Q-value for the action taken
            self.model.fit(state, target_f, epochs=1, verbose=0)  # Train the model
        if self.epsilon > self.epsilon_min:  # Decay exploration rate
            self.epsilon *= self.epsilon_decay  # Reduce exploration over time

    # Update the target model to match the main model
    def update_target_model(self):
        self.target_model.set_weights(self.model.get_weights())  # Copy weights from the main model to the target model

# Initialize the DQN agent
agent = DQNAgent(total_states, num_actions)  # Create agent with state and action space sizes

# Simulate the environment and train the agent
for episode in range(num_episodes):
    # Initialize the environment for the episode (random number of vehicles in each lane)
    vehicles_ew, vehicles_ns = random.randint(0, num_vehicles_per_lane), random.randint(0, num_vehicles_per_lane)
    traffic_phase = random.choice([0, 1])  # Random initial traffic phase (0 for NS green, 1 for EW green)
    state = get_state_index(vehicles_ew, vehicles_ns, traffic_phase)  # Convert state to index for neural network

    # Simulate the environment over multiple steps (episode)
    for step in range(max_steps_per_episode):
        # Choose an action using the DQN agent
        action = agent.act(np.array([state]))  # Convert state to array format for input to the model

        # Update environment state based on the chosen action
        prev_vehicles_ew, prev_vehicles_ns = vehicles_ew, vehicles_ns  # Store previous vehicle counts
        # Apply traffic simulation logic here (vehicle arrival/departure based on traffic phase)
        next_traffic_phase = action if action == 1 else traffic_phase  # Change phase if action is to switch
        next_state = get_state_index(vehicles_ew, vehicles_ns, next_traffic_phase)  # Get next state index

        # Calculate the reward based on state transition
        reward = calculate_reward(vehicles_ew, vehicles_ns, prev_vehicles_ew, prev_vehicles_ns, traffic_phase, action)

        # Check if the episode is done (e.g., after max steps)
        done = (step == max_steps_per_episode - 1)

        # Store experience in the agent's memory
        agent.remember(np.reshape(state, [1, total_states]), action, reward, np.reshape(next_state, [1, total_states]), done)

        # Move to the next state
        state = next_state
        traffic_phase = next_traffic_phase

        if done:
            agent.update_target_model()  # Update the target model after each episode
            break

    # Train the model after each episode
    agent.replay(batch_size=batch_size)
```
